```python
import pandas as pd

df = pd.read_csv("spotify_millsongdata.csv")   # or your actual filename
df['text'] = df['text'].str.lower().replace(r'^\w\s', ' ').replace(r'\n', ' ', regex=True)
```

```python
df['text'] = df['text'].str.lower().replace(r'^\w\s', ' ').replace(r'\n', ' ', regex=True)
```

```python
%pip install nltk
import nltk
from nltk.stem.porter import PorterStemmer

# Download required NLTK data
nltk.download('punkt')

stemmer = PorterStemmer()

def token(txt):
    token = nltk.word_tokenize(txt)
    a = [stemmer.stem(w) for w in token]
    return " ".join(a)
```

```python
token("you are beautiful, beauty")
```

```python
df ['text'] = df['text'].apply(lambda x: token(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidvector = TfidfVectorizer(analyzer='word',stop_words='english', max_features=5000)
matrix = tfidvector.fit_transform(df['text'])

# Compute similarity in chunks to avoid memory issues
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# For large datasets, compute similarities on-demand during recommendation
# Store matrix and vectorizer instead of the full similarity matrix
similarity = None  # We'll compute this on-demand in the recommendation function
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute similarity of the first song with all others (returns a 1D array)
# ensure TF-IDF matrix exists; if not, (re)create it from df['text']
if 'matrix' not in globals():
    tfidvector = TfidfVectorizer(analyzer='word', stop_words='english', max_features=5000)
    matrix = tfidvector.fit_transform(df['text'])

similar = cosine_similarity(matrix[0:1], matrix).flatten()
similar[0]
```

```python
df[df['song'] == 'Crying Over You']
```

```python
def recommendation(song_df):
	idx = df[df['song'] == song_df].index[0]
	# Compute similarity on-demand
	distance = sorted(list(enumerate(similar)), reverse=True, key=lambda x: x[1])
	song = []
	for s_id in distance[1:5]:
		song.append(df.iloc[s_id[0]].song)
	return song

recommendation('Crying Over You')
```
